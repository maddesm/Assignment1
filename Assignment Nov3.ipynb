```python
#IPRB
# Mendalian_inhertance : k (AA), m(Aa), n(aa): what we want ? those dominant phenotypical
# because each get 2 alleles from each individual we have to dive by 2
# AA * AA = 100 % all show dominant
# AA * Aa = 100 % all show dominant 
# AA * aa = 100 % all show dominant 
# the hetrozygus makes it tricky : 
# Aa * aa = 50 % show dominant 
# Aa * Aa = 75 % show dominant 

with open('rosalind_iprb.txt') as input_data:
        k, m, n = map(int, input_data.read().strip().split())

# k = homozygous dominant (AA) 
# m = Heterozygous (Aa)
# n = Homozygous recessive (aa)

def dominant_phenotype(k, m, n):
    total = k + m + n # population
    total_second = total - 1

    # Calculate the probability of each pairing
    prob_AA_AA = (k / total) * ((k - 1) / total_second)  # AA x AA
    prob_AA_Aa = (k / total) * (m / total_second) * 2       # AA x Aa
    prob_AA_aa = (k / total) * (n / total_second) * 2     # AA x aa
    prob_Aa_Aa = (m / total) * ((m - 1) / total_second) * 0.75  # Aa x Aa
    prob_Aa_aa = (m / total) * (n / total_second) * 2 * 0.5  # Aa x aa

    prob_dominant = (prob_AA_AA + prob_AA_Aa + prob_AA_aa +
                     prob_Aa_Aa + prob_Aa_aa)

    return prob_dominant


iprb = dominant_phenotype(k, m, n)
print(iprb)
```

```python
## FIB 
# Fibonacci sequece with rabits 
# solved with recursion becasue data set is too large
# base case would be 1 for each generation
with open('rosalind_fib.txt') as input_data:
        n, k = map(int, input_data.read().strip().split())

def fib(n, k):
    if n <= 2: # start from the second generation
        return 1 # base cases for first and second generations
    else:
        return fib(n - 1, k) + k * fib(n - 2, k)
print(fib(n, k))
```

```python
# SUBS
# Finding a Motif in DNA
from Bio.Seq import Seq

with open('rosalind_subs.txt') as input_data:
    s, t = input_data.readlines()
    s = s.strip()
    t = t.strip()

# Convert the strings to Seq objects so biopython will be able to read it
s_seq = Seq(s)
t_seq = Seq(t)

# Function to find all occurrences of the motif
def find_locations(s_seq, t_seq):
    locations = []  # Initialization (empty)
    for i in range(len(s_seq) - len(t_seq) + 1):
        if t_seq == s_seq[i: i + len(t_seq)]:
            locations.append(i + 1)
    return locations

locations = find_locations(s_seq, t_seq)

#locations as space-separated values (also I have to delete [])
print(" ".join(map(str, locations)))
```

```python
# Hamm (Hamming Distance)
# Counting Point Mutations
# give back an integer 

with open('rosalind_hamm.txt') as input_data:
    s, t = input_data.readlines()
    s = s.strip()
    t = t.strip()
    
def hamming_distance(s, t):
    distance = 0 
    assert len(s) == len(t) # checking if they're equal lenght
    for i in range(len(s)):
        if s[i] != t[i]:
            distance += 1
    return distance

print(hamming_distance(s, t))
```

```python
# ins(Insertion Sort)
# Problem wants us to count swaps.
with open('rosalind_ins.txt') as input_data:
    A = list(map(int, input_data.readlines()[1].strip().split()))
# I have to assign it as a list or it wont work 

def insertion_sort(A):
    swaps = 0  # Initialization
    for i in range(1, len(A)):
        while i > 0 and A[i] < A[i - 1]:
            A[i - 1], A[i] = A[i], A[i - 1] # swap elements A[i] and A[i-1]
            i -= 1
            swaps += 1  #update the swap counter
    return swaps
swaps = insertion_sort(A) # Get the number of swaps
print(swaps)
```

```python
# Mer(Merge Two Sorted Arrays)
# Merge Sort 
# Merges two sorted arrays A and B into a sorted array C

def merge_sort(A, B):
    C = []  # Initialize
    i, j = 0, 0
    while i < len(A) and j < len(B):
        if A[i] <= B[j]:
            C.append(A[i])
            i += 1
        else:
            C.append(B[j])
            j += 1
    # Add remaining elements from A OR B
    C += A[i:]
    C += B[j:]
    
    return C


if __name__ == '__main__':
    # Read input data
    with open("rosalind_mer.txt", "r") as f:
        n = int(f.readline().strip())
        A = list(map(int, f.readline().split()))
        m = int(f.readline().strip())
        B = list(map(int, f.readline().split()))

    C = merge_sort(A, B)
    print(" ".join(map(str, C)))
 
```

```python
# Par(2-Way Partition)

def partition(array):
    new_array, left_array, right_array = [], [], []
    new_array.append(array[0])
    for i in range(1, len(array)):
        if array[i] <= array[0]:
            left_array.append(array[i])
        else:
            right_array.append(array[i])
    new_array = left_array + new_array + right_array
    return new_array

if __name__ == '__main__':
    with open("rosalind_par.txt", "r") as f:
        n = int(f.readline().strip())
        A = [int(i) for i in f.readline().strip().split(" ")]
    B = partition(A)
    for i in B:
        print(i, end=" ")
    print()
```

```python
#Par3(3-Way Partition)
def partition3(array):
    new_array, left_array, right_array = [], [], []
    new_array.append(array[0])
    for i in range(1, len(array)):
        if array[i] < array[0]:
            left_array.append(array[i])
        elif array[i] == array[0]:
            new_array.append(array[i])
        else:
            right_array.append(array[i])
    new_array = left_array + new_array + right_array
    return new_array    

if __name__ == '__main__':
    with open("rosalind_par3.txt", "r") as f:
        n = int(f.readline().strip())
        A = [int(i) for i in f.readline().strip().split(" ")]
    results = partition3(A)
    for r in results:
        print(r, end=" ")
    print()
```

```python
# QS(Quick Sort)

def partition(arr, low, high):
    # Select pivot element
    i = low - 1
    for j in range(low, high):
        if arr[j] <= arr[high]:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]
    arr[i + 1], arr[high] = arr[high], arr[i + 1]
    return i + 1

def quick_sort(arr, low, high):
    if low < high:
        # Partitioning index
        pi = partition(arr, low, high)
        # Recursively sort the subarrays
        quick_sort(arr, low, pi - 1)
        quick_sort(arr, pi + 1, high)
    return arr

if __name__ == "__main__":
    # Read input data from file
    with open("rosalind_qs.txt", "r") as f:
        n = int(f.readline().strip())  # Size of the array
        A = [int(i) for i in f.readline().strip().split()]  # The array to be sorted

    # Sort the array using Quick Sort
    sortedA = quick_sort(A, 0, n - 1)

    # Output the sorted array
    print(" ".join(map(str, sortedA)))
```

```python
#HS (Heap Sort)
def heapify(arr, n, i): 
    largest = i
    l = 2 * i + 1
    r = 2 * i + 2

    # If left child is larger than root
    if l < n and arr[l] > arr[largest]: 
        largest = l

    # If right child is larger than largest so far
    if r < n and arr[r] > arr[largest]: 
        largest = r

    # If largest is not root
    if largest != i: 
        arr[i], arr[largest] = arr[largest], arr[i]
        heapify(arr, n, largest)  # Recursively heapify the affected subtree
  
def buildHeap(arr, n):
    # Start from the last non-leaf node and heapify each subtree
    for i in range(n // 2 - 1, -1, -1): 
        heapify(arr, n, i)

if __name__ == '__main__': 
    with open("rosalind_hs.txt", "r") as f:
        n = int(f.readline().strip())  # Size of the array
        arr = [int(i) for i in f.readline().strip().split()]  # The array to be sorted

    # Build a max heap
    buildHeap(arr, n)

    # Extract elements from the heap one by one
    for i in range(n - 1, 0, -1):
        arr[0], arr[i] = arr[i], arr[0]  # Move current root to the end
        heapify(arr, i, 0)  # Heapify the reduced heap

    # Output the sorted array
    print(" ".join(map(str, arr)))
```

```python
# (Ps) Partial Sort 
def heapify(arr, n, i):
    smallest = i  # Start with the root as the smallest
    l = 2 * i + 1  # Left child index
    r = 2 * i + 2  # Right child index
    
    # Check if left child is smaller than root
    if l < n and arr[l] < arr[smallest]:
        smallest = l
    
    # Check if right child is smaller than root
    if r < n and arr[r] < arr[smallest]:
        smallest = r
    
    # If the smallest is not root, swap them and heapify the affected subtree
    if smallest != i:
        arr[i], arr[smallest] = arr[smallest], arr[i]
        heapify(arr, n, smallest)

def buildHeap(arr, n):
    # Build the min-heap (start from the last non-leaf node)
    for i in range(n//2 - 1, -1, -1):
        heapify(arr, n, i)

if __name__ == '__main__':
    with open("rosalind_ps.txt", "r") as f:
        n = int(f.readline().strip())  # Read the number of elements
        arr = [int(i) for i in f.readline().strip().split()]  # Read the array
        k = int(f.readline().strip())  # Read the number of elements to extract
    
    # Perform the partial sort by extracting the smallest element k times
    for i in range(k):
        buildHeap(arr, len(arr))  # Build a heap for the remaining elements
        print(arr[0], end=" ")  # Print the smallest element (root of the heap)
        arr[0] = arr[-1]  # Replace the root with the last element
        arr.pop()  # Remove the last element
    print()
```
